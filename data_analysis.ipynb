```python
reg_seasonal_stats = df.drop(columns=["date","home","away"]).loc[df["type"] != "playoff"]
reg_seasonal_stats = reg_seasonal_stats.groupby(["season","playerid"],as_index=False).agg({
                                                    "player":lambda x: x.iloc[0],
                                                   "team":[lambda x: list(set(x)),lambda x: len(set(x))],
                                                   "PTS":["mean","sum"],
                                                   "REB":["mean","sum"],
                                                   "AST":["mean","sum"],
                                                   "STL":["mean","sum"],
                                                   "BLK":["mean","sum"],
                                                   "gameid": lambda x:len(x)
                                                   })
reg_seasonal_stats.rename(columns = {"gameid":"games played"},inplace=True)
```

```python
#Show point leaders by season -- last bit to reorder by season
pts_leaders = reg_seasonal_stats.sort_values(("PTS","mean"),ascending=False).drop_duplicates(("season",""),keep="first").sort_values("season")
pts_leaders

```

```python
# Find all triple doubles in the regular season
traditional_reg = df.loc[df["type"] != "playoff"]
off_stats = traditional_reg[["PTS","REB","AST","STL","BLK"]]
triple_doubles = traditional_reg[(off_stats[off_stats.columns]>10).sum(1) > 2]
#triple_doubles.to_csv("triple_doubles.csv",index = False)
```

```python
# Count the number of triple doubles
td_leaders = triple_doubles.groupby("player",).apply(lambda x: x)["player"].value_counts()
td_leaders
```

```python
td_top_6 = td_leaders.index[0:5]
```

```python
# Show a specific players td's
triple_doubles.groupby("player").apply(lambda x: x).loc[td_top_6]
```
